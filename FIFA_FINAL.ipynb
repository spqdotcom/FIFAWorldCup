```python
import pandas as pd

import numpy as np
```

```python
WorldCupMatches = 'WorldCupMatches.csv'
WorldCupPlayers = 'WorldCupPlayers.csv'
WorldCups = 'WorldCups.csv'
```

```python
df_mat = pd.read_csv(WorldCupMatches)
df_pla = pd.read_csv(WorldCupPlayers)
df_wc = pd.read_csv(WorldCups)
```

```python
#Check percentage of empty cells per column
for x in df_mat.columns:
    print(f"The percentage of empty cells in column {x} is {round(df_mat[x].isna().sum()/len(df_mat[x])*100,2)} %.")
```

```python
#Check percentage of empty cells per column
for x in df_pla.columns:
    print(f"The percentage of empty cells in column {x} is {round(df_pla[x].isna().sum()/len(df_pla[x])*100,2)} %.")
```

```python
#Check percentage of empty cells per column
for x in df_wc.columns:
    print(f"The percentage of empty cells in column {x} is {round(df_wc[x].isna().sum()/len(df_wc[x])*100,2)} %.")
```

```python
#Deleted empty rows
df_mat = df_mat.dropna(how='all')
df_mat
```

```python
#Delete duplicate rows
df_mat = df_mat.drop_duplicates()
df_mat
```

```python
#Delete duplicate rows
df_pla = df_pla.drop_duplicates()
df_pla
```

```python
#Delete duplicate rows
df_wc = df_wc.drop_duplicates()
df_wc
```

```python
#Use the fillna() function to replace NaN with 0 and then convert to int
df_mat['Attendance'] = df_mat['Attendance'].fillna(0).astype(int)

#Check the result
print(df_mat['Attendance'].head())
```

```python
#Change data type of each column
df_mat['Year'] = df_mat['Year'].astype('int64')
df_mat['RoundID'] = df_mat['RoundID'].astype('int64')
df_mat['MatchID'] = df_mat['MatchID'].astype('int64')
df_mat['Attendance'] = df_mat['Attendance'].astype('int64')
```

```python
#Rename columns
df_mat.rename(columns={'Year': 'year_id'}, inplace=True)
print(df_mat.head())

#Rename columns
df_wc.rename(columns={'Year': 'year_id'}, inplace=True)
print(df_wc.head())
```

```python
#Sort columns
first_columns = ['MatchID', 'year_id', 'RoundID']
other_columns = [col for col in df_mat.columns if col not in first_columns]
df_mat = df_mat[first_columns + other_columns]
print(df_mat.head())
```

```python
#Deletes columns
df_pla = df_pla.drop(columns=['Shirt Number', 'Position', 'Event'])
df_pla.dtypes
```

```python
#Add index
df_pla['player_id'] = range(1, len(df_pla) + 1)
print(df_pla.head())
```

```python
#Sort columns
columns = ['player_id'] + [col for col in df_pla if col != 'player_id']
df_pla = df_pla[columns]
print(df_pla.head())
```
